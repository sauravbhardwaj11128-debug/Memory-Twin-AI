```python
embedding_model = SentenceTransformer(
    "all-MiniLM-L6-v2"
)


client = chromadb.Client()

memory_db = client.create_collection(
    name="memory_twin"
)
```

```python
memories = [
    "I believe career growth is more important than quick money.",
    "I prefer learning new technology and experimenting.",
    "I usually take decisions after thinking about long term effects.",
    "I value honesty and consistency."
]


for i,memory in enumerate(memories):

    vector = embedding_model.encode(memory).tolist()

    memory_db.add(
        ids=[str(i)],
        documents=[memory],
        embeddings=[vector]
    )

print("Memory stored")
```

```python
ARMORIQ_API_KEY = "ak_live_1ca54cbfe24aaf748af54e4bceb5602d73e8eb51c66c3f347195a1c04abdb86b"


def armor_policy_check(query):

    blocked_words = [
        "password",
        "secret",
        "private key"
    ]

    for word in blocked_words:
        if word in query.lower():
            return False

    return True
```

```python
audit_logs=[]


def log_action(query,response):

    audit_logs.append(
        {
            "time":str(datetime.now()),
            "query":query,
            "response":response
        }
    )
```

```python
def retrieve_memory(query):

    q_vector = embedding_model.encode(query).tolist()


    result = memory_db.query(
        query_embeddings=[q_vector],
        n_results=3
    )


    return result["documents"][0]
```

```python
def memory_twin_agent(query):


    # ArmorIQ check

    allowed = armor_policy_check(query)


    if not allowed:
        return "Blocked by ArmorIQ Security Policy"



    memories = retrieve_memory(query)


    prompt=f"""

You are a personal memory twin.

Answer like the person whose memories are given.

Memories:
{memories}


User question:
{query}


Give helpful advice based on personality.
Do not claim you are the real person.

"""


    response = model.generate_content(prompt)


    answer=response.text


    log_action(query,answer)


    return answer
```

```python
question = """
Should I choose a risky career option or a safe job?
"""


print(
    memory_twin_agent(question)
)
```
